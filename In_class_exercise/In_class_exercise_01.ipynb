(2) Write a Python program to count the number of characters in a string. (4 points)

Sample String : 'google.com‘, 

Expected Result : {'o': 3, 'g': 2, '.': 1, 'e': 1, 'l': 1, 'm': 1, 'c': 1}

```python
# write your answer here
test_str = input("Please enter a string: ")
dict_str = {}
for i in test_str:
  if i in dict_str:
    dict_str[i] += 1
  else:
    dict_str[i] = 1
sorted_str = sorted(dict_str.items(), key=lambda kv:kv[1], reverse = True)
print(sorted_str)
```

(4) Write a Python program to get a string from a given string where all occurrences of its first char have been changed to '\$', except the first char itself. (4 points)

Sample String : 'restart‘, 

Expected Result : 'resta\$t'

```python
# write your answer here
def change_char(str1):  
  char = str1[0]  
  str1 = str1.replace(char, '$')  
  str1 = char + str1[1:]  
  return str1  
print(change_char('restart'))
```

(6) Write a Python program to concatenate following dictionaries to create a new one. (4 points)

Sample Dictionary : dic1={1:10, 2:20}, dic2={3:30, 4:40}, dic3={5:50,6:60}, 

Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}


```python
# write your answer here
dic1 = {1:10,2:20}
dic2 = {3:30,4:40}
dic3 = {5:50,6:60}
new_dic = {}
for dictionary in (dic1,dic2,dic3):
  new_dic.update(dictionary)
print(new_dic)
```

(10) Write a Python program to find the duplicate elements in a given array of integers. Return -1 If there are no such elements. (4 points)

```python
def duplicate(arr):
  arr_set = set()
  no_duplicate = -1
  for i in range(len(arr)):
    if arr[i] in arr_set:
      return arr[i]
    else:
      arr_set.add(arr[i])
  return no_duplicate

arr_1 = [1,2,3]
arr_2 = [1,2,2]

print("The array 1 is: " + str(arr_1))
print("The array 2 is: " + str(arr_2))

print(duplicate(arr_1))
print(duplicate(arr_2))
```

(11) Write a Python program to select all the Sundays of a specified year. (4 extra points)

```python
# write your answer here
from datetime import date, timedelta
year = 2021
first_d = date(year, 1, 1)                    
first_d = first_d + timedelta(days = 6 - first_d.weekday())  
while first_d.year == year:
  print(first_d)
  first_d = first_d + timedelta(days = 7)
```
