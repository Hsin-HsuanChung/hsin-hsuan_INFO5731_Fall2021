```python
# open file
with open('review.txt') as file:
    txt = file.read()

df = pd.DataFrame(txt.split("\n"), columns =['Text Data'])
df.to_csv('O_result.csv')
```

```python
# clean text
import urllib.request
import re
import requests
import pandas as pd

#Lower casing
df['Text Data'] = df['Text Data'].apply(lambda x: " ".join(x.lower() for x in x.split()))

#Removing Punctuations
import string
df['Text Data'] = df['Text Data'].str.replace('[{}]'.format(string.punctuation), '')

#Removing Stopwords
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop=stopwords.words('english')
df['Text Data'] = df['Text Data'].apply(lambda x:" ".join(x for x in x.split() if x not in stop))

#frequent word removal
freq=pd.Series(' '.join(df['Text Data']).split()).value_counts()[:10]
freq=list(freq.index)
df['Text Data'] = df['Text Data'].apply(lambda x:" ".join(x for x in x.split() if x not in freq))

#rare word removal
freq=pd.Series(' '.join(df['Text Data']).split()).value_counts()[-10:]
freq=list(freq.index)
df['Text Data'] = df['Text Data'].apply(lambda x:" ".join(x for x in x.split() if x not in freq))

#Spelling Correction
from textblob import TextBlob
df['Text Data'] = df['Text Data'].apply(lambda x: str(TextBlob(x).correct()))

#stemming
from nltk.stem import PorterStemmer
st=PorterStemmer()
df['Text Data'] = df['Text Data'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))

#Lemmetization
from textblob import Word
# import nltk
nltk.download('wordnet')
df['Text Data'] = df['Text Data'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))

#Tokenization
import nltk
nltk.download('punkt')
df['Text Data'] = df['Text Data'].apply(lambda x: TextBlob(x).words)

df.to_csv('clean_text.csv')
```

```python
# 1-1 Count the frequency of all the N-grams (N=3)
with open('clean_text.csv', encoding="utf8", errors='ignore') as f:
    sentence = f.read()

import itertools
import collections
from nltk import ngrams
from nltk.tokenize import word_tokenize

words_sentence = []
for sentence in df['Text Data']:
    words_sentence.append(sentence)
aftercleaning = [x for x in words_sentence if x != []]
iterations = list(itertools.chain.from_iterable(aftercleaning))

# Count the frequency of all the N-grams (N=3)
tri_grams = nltk.trigrams(iterations)
frequency_dist = nltk.FreqDist(tri_grams)
tri_grams_dict = dict(frequency_dist)
for word in tri_grams_dict:
    print( str(word) + ':'+ str(frequency_dist[word]))
```

```python
# 1-2 Calculate the probabilities for all the bigrams in the dataset
from collections import Counter
bi_grams = nltk.bigrams(iterations)
frequency_dist = nltk.FreqDist(bi_grams)
bi_grams_dict = dict(frequency_dist)
for word in bi_grams_dict:
    print( str(word) + ':' + str(bi_grams_dict[word] / iterations.count(word[0])))
```
