```python
# open file
with open('01-05-1  Adams v Tanner.txt') as file:
    txt = file.read()
txt.replace('\r',' ')
txt.replace('\n',' ')
```

```python
##(1)##
# Number of sentences
print("Number of sentences:", txt.count('.'))

# Number of words
print('Number of words:', txt.count(' '))

# Number of characters
characters = [char for char in list(txt) if char != ' ']
print('Number of charcters: ', len(list(txt)))

# Average word length
words = txt.split()
average = sum(len(word) for word in words) / len(words)
print('Average word length: ',average)

# Number of Stopwords
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
print('Number of stopwords:', len([word for word in txt.split() if word in stopwords.words('english')]))

# Number of Special Characters
spl_chars = '!@#$^&*'
tot = 0
for char in spl_chars:
    count = len([word for word in txt.split(' ') if word.startswith(char)])
    tot += count
print('Number of special characters:', tot)

# Number of Numerics
numerics = len([x for x in list(txt) if x.isnumeric()])
print('Number of numerics:', numerics)

# Number of Uppercase Words
uppercase_words = len([word for word in txt.split() if word.isupper()])
print('Number of uppercase words:', uppercase_words)
```

```python
##(2)##
import urllib.request
import re
import requests
import pandas as pd
text_data=[]
url= urllib.request.urlopen("https://raw.githubusercontent.com/unt-iialab/info5731_spring2021/main/class_exercises/01-05-1%20%20Adams%20v%20Tanner.txt")
for text in url:
    decode_text=text.decode("utf-8")
    clean_text=decode_text.replace('\r\n','')
    if clean_text:
        text_data.append(clean_text)

df=pd.DataFrame(text_data,columns=['Text Data'])

#Lower casing
df['Text Data'] = df['Text Data'].apply(lambda x: " ".join(x.lower() for x in x.split()))

#Removing Punctuations
import string
df['Text Data'] = df['Text Data'].str.replace('[{}]'.format(string.punctuation), '')

#Removing Stopwords
# from nltk.corpus import stopwords
# import nltk
nltk.download('stopwords')
stop=stopwords.words('english')
df['Text Data'] = df['Text Data'].apply(lambda x:" ".join(x for x in x.split() if x not in stop))

#frequent word removal
freq=pd.Series(' '.join(df['Text Data']).split()).value_counts()[:10]
freq=list(freq.index)
df['Text Data'] = df['Text Data'].apply(lambda x:" ".join(x for x in x.split() if x not in freq))

#rare word removal
freq=pd.Series(' '.join(df['Text Data']).split()).value_counts()[-10:]
freq=list(freq.index)
df['Text Data'] = df['Text Data'].apply(lambda x:" ".join(x for x in x.split() if x not in freq))

#Spelling Correction
from textblob import TextBlob
df['Text Data'] = df['Text Data'].apply(lambda x: str(TextBlob(x).correct()))

#stemming
from nltk.stem import PorterStemmer
st=PorterStemmer()
df['Text Data'] = df['Text Data'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))

#Lemmetization
from textblob import Word
# import nltk
nltk.download('wordnet')
df['Text Data'] = df['Text Data'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))

#Tokenization
import nltk
nltk.download('punkt')
df['Text Data'] = df['Text Data'].apply(lambda x: TextBlob(x).words)
df
```

```python
##(3)##
# Save to .csv file
import pandas as pd
# dt_fr = pd.DataFrame({'Sentences' :txt.split('.')})
df.to_csv('clean_text.csv')
df
```

```python
##(4-1)##
import numpy as np
import pandas as pd
# Calculate the term frequency of all the terms
term_freq = (df['Text Data'].apply(lambda x: pd.value_counts(str(x).split(" "))).sum(axis = 0).reset_index())
term_freq.columns = ['words','tf']
term_freq.tf = np.where(term_freq.tf != 0, 1+np.log(term_freq.tf), 0)
term_freq
```

```python
##(4-2)##
#top 10 1-gram, top 10 2-grams, and top 10 3-grams terms
from textblob import TextBlob
from textblob import Word
joint_sent = [" ".join(sublist) for sublist in df['Text Data']]
joint_text = " ".join (joint_sent) 
#print(joint_text)
  #print(joined_sent)
def n_gram(text, m):
    blob = TextBlob(text)
    n_gram = blob.ngrams(n=m)
    print("Top 10 "+ str(m) + "_gram:", n_gram[:11])

n_gram(joint_text, 1)
print()
n_gram(joint_text, 2)
print()
n_gram(joint_text, 3)
```
